```python
#Data Cleaning Steps 
#Checking for the duplicates 
a = df[df.duplicated()]
a
```

```python
#removal of the duplicates
a = df.drop_duplicates()
a
```

```python
#removing the unwanted columns
column_name = ['restecg','thal','slope','ca']
b = a.drop(columns=column_name)
b
```

```python
#checking for the null values
c = b.isnull().sum()
c
```

```python
#replacing the mean,median, mode with missing values
column_to_replace = ['chol','oldpeak','thalach','exang']
for columns in column_to_replace:
    mean_value = b[columns].mean()
    b[columns].fillna(mean_value,inplace=True)  
b
```

```python
#checking for the null values
c = b.isnull().sum()
c
```

```python
#checking for the skewness
skewness = {}
column_to_transform =['age','sex','cp','trestbps','chol','fbs','thalach','exang','oldpeak']
for columns in column_to_transform:
    skewness_dict = b[columns].skew()
    skewness[columns] = skewness_dict
for columns,skewness_dict in skewness.items():
    print(f"{columns}:{skewness_dict}")
```

```python
#skewness reduction 
import numpy as np
column_to_change =['age','sex','thalach']
for columns in column_to_change:
    b[columns] = np.log1p(b[columns])
```
