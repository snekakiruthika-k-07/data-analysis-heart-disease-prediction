```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset
df = pd.read_csv("heart disease prediction.csv")
df
```

```python
#displaying the columns 
df.columns
```

```python
#visualization techinques 
#barchart 
columns_to_change =['sex', 'target']
df[columns_to_change].value_counts().plot(kind='bar',color='green')
plt.title("HEART DISEASE PREDICTION")
plt.show()
```

```python
#piechart
df.plot(kind='pie',x='sex',y='target')
plt.show()
```

```python
#histogram
sns.histplot(data=df, x='age', y='target', hue='target', palette='Set1')
plt.xlabel("AGE")
plt.ylabel("TARGET")
plt.title("AGE VS TARGET")
plt.show()

```

```python
#areachart 
df.plot(kind='area',x='chol',y='age',color='crimson')
plt.xlabel("CHOLESTROL")
plt.ylabel("AGE")
plt.show()
```

```python
#scatterplot
sns.scatterplot(data=df, x='trestbps', y='target', hue='target', palette='Set1')
plt.title("HEART DISEASE PREDICTION")
plt.show()
```

```python
#boxplot
sns.boxplot(df['age'])
plt.title("HEART DISEASE PREDICTION")
plt.show()
```

```python
#heatmap 
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```
